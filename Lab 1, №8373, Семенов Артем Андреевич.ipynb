Stroke Prediction Dataset

Предметная область: медицина  
Характер данных: реальные  
Источник данных: https://www.kaggle.com/datasets/fedesoriano/stroke-prediction-dataset?resource=download&select=healthcare-dataset-stroke-data.csv

Атрибуты, представленные в датасете:

1) id - уникальный идентификатор пациента(int)
2) gender - пол, 0 - мужской, 1 - женский (int)
3) age - возраст (int)
4) hypertension - 0 (отсутствие гипертонии) или 1 (наличие гипертонии) (bool)
5) heart_disease - 0 (отсутствие заболеваний сердца) или 1 (их наличие) (bool)
6) ever_married - был ли пациент когда-либо женат, 0 - нет, 1 - да (bool)
7) work_type - статус занятости (int)  
    1 - "children",  
    2 - "Never_worked",  
    3 - "Govt_job",  
    4 - "Self-employed",  
    5 - "Private"
8) residence_type - место проживания, 0 - деревня, 1 - город (int)
9) avg_glucose_level - средний уровень глюкозы в крови (float)
10) bmi - индекс массы тела (float)
11) smoking_status - информация о курении (int)  
    0 - "never smoked",  
    1 - "formerly smoked",  
    2 - "smokes"  
    Unknown*
12) stroke - 1 (у пациента уже был инсульт) или 0 (не было) (bool)

*Примечание: "Unknown" в поле smoking_status означает, что информация недоступна для данного пациента.

Описание задачи анализа:  
По данным Всемирной организации здравоохранения (ВОЗ) инсульт является второй по значимости причиной смерти в мире, на его долю приходится примерно 11% от общего числа смертей.  
Этот набор данных используется для прогнозирования вероятности инсульта у пациента на основе входных параметров, таких как пол, возраст, различные заболевания, статус курения и др.

Определение корреляции

```python
# Возраст и наличие гипертонии имеют среднюю корреляцию 
# Чем больше возраст, тем более вероятно наличие гипертонии у пациента

print(dataset['age'].corr(dataset['hypertension']))
print(dataset['age'].corr(dataset['hypertension'], method='spearman'))
#dataset.plot.scatter(x = 'age', y = 'hypertension')
```

```python
# Возраст и наличие заболеваний сердца имеют среднюю корреляцию 
# Чем больше возраст, тем более вероятно наличие заболеваний сердца у пациента

print(dataset['age'].corr(dataset['heart_disease']))
print(dataset['age'].corr(dataset['heart_disease'], method='spearman'))
#dataset.plot.scatter(x = 'age', y = 'heart_disease')
```

```python
# Возраст и наличие брака имеют среднюю корреляцию 
# Чем больше возраст, тем более вероятно наличие брака

print(dataset['age'].corr(dataset['ever_married']))
print(dataset['age'].corr(dataset['ever_married'], method='spearman'))
#dataset.plot.scatter(x = 'age', y = 'ever_married')
```

```python
# Возраст и уровень глюкозы в крови имеют слабую корреляцию
# Чем больше возраст, тем более вероятно наличие у пациента высокого уровня глюкозы

print(dataset['age'].corr(dataset['avg_glucose_level']))
print(dataset['age'].corr(dataset['avg_glucose_level'], method='spearman'))
#dataset.plot.scatter(x = 'age', y = 'avg_glucose_level')
```

```python
# Возраст и индекс массы тела имеют среднюю корреляцию
# ИМС достигает своего максимума в середине жизненного цикла, после чего начинает плавно снижаться

print(dataset['age'].corr(dataset['bmi']))
print(dataset['age'].corr(dataset['bmi'], method='spearman'))
#dataset.plot.scatter(x = 'age', y = 'bmi')
```
